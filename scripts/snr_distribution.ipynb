```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
snr_5e4 = np.loadtxt("../data/sorted_redbook_50000.0.txt")
snr_5e5 = np.loadtxt("../data/sorted_redbook_500000.0.txt")
```

```python
counts5e4, bins5e4 = np.histogram(snr_5e4[:,0], bins=50)
counts5e5, bins5e5 = np.histogram(snr_5e5[:,0], bins=50)
#plt.stairs(counts5e3, bins5e3, fill=True)
plt.stairs(counts5e4, bins5e4, fill=True, alpha=0.6, label='M=50000')
plt.stairs(counts5e5, bins5e5, fill=True, alpha=0.4, label='M=500000')
plt.legend()
plt.xlabel('SNR')
plt.ylabel('count')
#plt.show()
plt.savefig('../plots/few_snr_estimates')
```

```python
snr_AAK = np.loadtxt("../data/sorted_redbook_AAK_2000.txt")
snr_FEW = np.loadtxt("../data/sorted_redbook_FEW_2000.txt")
snr_FEW_low_e = np.loadtxt("../data/sorted_redbook_FEW_low_ecc.txt")
snr_FEW_tdi1 = np.loadtxt("../data/sorted_redbook_FEW_low_ecc_TDI1.txt")
```

```python
countsAAK, binsAAK = np.histogram(snr_AAK[:,0], bins=50)
countsFEW, binsFEW = np.histogram(snr_FEW[:,0], bins=50)
countsFEW_e0, binsFEW_e0 = np.histogram(snr_FEW_low_e[:,0], bins=50)
countsFEW_e0_double, binsFEW_e0_double = np.histogram(2.*snr_FEW_low_e[:,0], bins=50)
countsFEW_tdi1, binsFEW_tdi1 = np.histogram(snr_FEW_tdi1[:,0], bins=50)
#plt.stairs(counts5e3, bins5e3, fill=True)
# plt.stairs(countsAAK, binsAAK, fill=True, alpha=0.7, label='AAK')
# plt.stairs(countsFEW, binsFEW, fill=True, alpha=0.5, label='FEW')
plt.stairs(countsFEW_e0, binsFEW_e0, fill=True, alpha=0.8, label='FEW TDI2')
plt.stairs(countsFEW_e0_double, binsFEW_e0_double, fill=True, alpha=0.5, label='FEW TDI2 doubled')
plt.stairs(countsFEW_tdi1, binsFEW_tdi1, label='FEW TDI1')
plt.legend()
plt.xlabel('SNR')
plt.ylabel('count')
plt.show()
#plt.savefig('few_snr_estimates_aak_vs_few_e0_tdi1')
```

```python
snr_FEW_tdi1 = np.loadtxt("../data/sorted_redbook_FEW_low_ecc_TDI1_v2.txt")
snr_AAK = np.loadtxt("../data/sorted_redbook_AAK_low_ecc.txt")
snr_phenoms = np.loadtxt("../data/SNRs4M=500000-TDI2AET.dat")
snr_SBHB = snr_phenoms[:,0]
snr_IMRPhenomD = snr_phenoms[:,1]
snr_IMRPhenomHM = snr_phenoms[:,2]
```
